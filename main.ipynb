```python
import yfinance as yf
import pandas as pd
```

```python
c_df = pd.read_csv('./S&P 500 Historical Components & Changes(03-10-2025).csv',index_col=0)
```

```python
c_df
```

```python
c_df.index = pd.to_datetime(c_df.index)
```

```python
c_df
```

```python
c_df.iloc[-1].tickers.split(',')
```

```python
def get_tickers_date(c_df, date):
    if date >= c_df.index[-1]:
        ticks = c_df.iloc[-1].tickers
    else:
        ticks = c_df.loc[date:].iloc[0].tickers
    return ticks.split(',')
```

```python
s_p_tickers = c_df.iloc[-1].tickers.split(',')
```

```python
df = yf.download(s_p_tickers,start='2010-01-01')['Close']
```

```python
df
```

```python
monthly_prices = df.resample('ME').last()
```

```python
momentum = monthly_prices.pct_change(12).shift(1)
```

```python
momentum
```

```python
portfolio = pd.DataFrame(index=monthly_prices.index,columns=monthly_prices.columns)
```

```python
portfolio
```

```python
for date in momentum.index[12:]:
    pf_date = momentum.loc[date]
    rel_tickers = get_tickers_date(c_df,date)
    pf_date = pf_date[pf_date.index.isin(rel_tickers)]
    top20 = pf_date.nlargest(20).index
    portfolio.loc[date,top20] = 1/20
```

```python
date
```

```python
top20
```

```python
portfolio.loc[date,top20]
```

```python
returns = monthly_prices.pct_change().shift(-1)
```

```python
strategy_returns =(portfolio * returns).sum(axis=1)
```

```python
(1+strategy_returns).cumprod().plot(title='cross_sectional_strategy return')
```
